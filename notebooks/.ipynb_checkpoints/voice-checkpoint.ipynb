```python
import pandas as pd
import checklist
from allennlp_models.pretrained import load_predictor
from allennlp_models.pretrained import get_pretrained_models
from checklist.editor import Editor
from checklist.perturb import Perturb
from checklist.test_types import MFT, INV, DIR
from checklist.expect import Expect
from checklist.pred_wrapper import PredictorWrapper
```

```python

editor = Editor()
```

```python
def make_list_shoot_passive(amount_per_format):
    smacked_passive_list = []
    smacked_passive = editor.template("{first_name1} was shot by {first_name2}.",
                                    meta=True, 
                                    nsamples=amount_per_format, 
                                    remove_duplicates=True, 
                                    seed=18)
    smacked_passive_list.append(smacked_passive['data'])
    smacked_passive_list = [x for xs in smacked_passive_list for x in xs]
    return smacked_passive_list

def make_list_shoot_active(amount_per_format):
    smacked_passive_list = []
    smacked_passive = editor.template("{first_name1} shoots {first_name2}.",
                                    meta=True, 
                                    nsamples=amount_per_format, 
                                    remove_duplicates=True, 
                                    seed=18)
    smacked_passive_list.append(smacked_passive['data'])
    smacked_passive_list = [x for xs in smacked_passive_list for x in xs]
    return smacked_passive_list
```

```python
def get_tags_v0(sentence):
    output = srl_predictor.predict(sentence)
    #print(output['verbs'])
    if len(output['verbs']) >= 1:
        tags = output['verbs'][0]['tags']#['verb']#[0]['tags']
    else:
        tags = []
    return tags
        
def get_tags_v1(sentence):
    output = srl_predictor.predict(sentence)
    #print(output['verbs'])
    if len(output['verbs']) >= 1:
        tags = output['verbs'][1]['tags']#['verb']#[0]['tags']
    else:
        tags = []
    return tags
        
def get_descr_v0(sentence):
    output = srl_predictor.predict(sentence)
    if len(output['verbs']) >= 1:
        descr = output['verbs'][0]['description']#['verb']#[0]['tags']
    else:
        descr = []
    return descr

def get_descr_v1(sentence):
    output = srl_predictor.predict(sentence)
    if len(output['verbs']) >= 1:
        descr = output['verbs'][1]['description']#['verb']#[0]['tags']
    else:
        descr = []
    return descr

def find_final_min_one_arg(list_of_arg):
    return list_of_arg[-2]

def find_first_arg(list_of_arg):
    return list_of_arg[0]
```

```python
active = make_list_shoot_active(10)
passive = make_list_shoot_passive(10)

df = pd.DataFrame(active)
df[1] = passive
df['tags_active'] = df[0].apply(get_tags_v0)
df['descr_act'] = df[0].apply(get_descr_v0)
df['tags_pass'] = df[1].apply(get_tags_v1)
df['descr_pass'] = df[1].apply(get_descr_v1)
df
```
